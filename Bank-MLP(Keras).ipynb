```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./bank.csv", sep=";")
```

```python
LE = LabelEncoder()
```

```python
df['job_code'] = LE.fit_transform(df['job'])
df['marital_code'] = LE.fit_transform(df['marital'])
df['education_code'] = LE.fit_transform(df['education'])
df['housing_code'] = LE.fit_transform(df['housing'])
df['loan_code'] = LE.fit_transform(df['loan'])
df['contact_code'] = LE.fit_transform(df['contact'])
df['poutcome_code'] = LE.fit_transform(df['poutcome'])
df['subscribed'] = LE.fit_transform(df['y'])
```

```python
df=df.drop(['job','marital', 'balance', 'education','housing','loan','contact','poutcome','y','default', 'balance','day', 'month', 'duration'],axis=1)
```

```python
df.head()
```

```python
X=df.drop('subscribed',axis=1)
y= df['subscribed']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=0)
```

```python
scalar= StandardScaler()
scalar.fit(X_train)
```

```python
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

```python
model = Sequential()
```

```python
model.add(Dense(14, input_dim=11, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
```

```python
print('MSE:{}'.format(score))
```
